### 題1 我是不是該去看醫生
這個是個簡單的機率問題。假定有個醫學界阿宅檢定號稱通過考試的阿宅有95%的機率是個阿宅，而當前人口有99.9%是正常人。
請問
1.）隨便抓一個人去檢驗而出現陽性反應的機率有多少？
$${ P}(B)={ P}(A){ P}(B\mid A)+{ P}(A^ c){ P}(B\mid A^ c) =0.001 \cdot 0.95 + 0.999 \cdot 0.05 =0.0509.$$

```python
pa = 0.001
pbga = 0.95
pac = 1-pa
pbgac = 0.05
print "Total probability of P(B) is " + \
        str(0.001*0.95 + 0.05* 0.999)
```

你被檢出，但妳趁的事

$${ P}(A\mid B)=\frac{{ P}(A){ P}(B\mid A)}{{ P}(B)} =\frac{0.001 \cdot 0.95}{0.0509} \approx 0.01866.$$
